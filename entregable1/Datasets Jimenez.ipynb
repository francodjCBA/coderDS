### <b>DataSet 1:</b> DNRPA - Inscripción Automóviles enero 2022 - Argentina

El primer set de datos seleccionado es sobre la inscripción de automóviles en la <b>Dirección Nacional de los Registros Nacionales de la Propiedad del Automotor (DNRPA)</b> en la República Argentina desde el 01 de enero de 2022 al 31 de enero de 2022.

En este dataset las variables más interesantes son:
  
  - Caracteristicas de vehiculos: Marca, modelo, año.
  - Caracteristicas de personas que realizan el registro: Edad, Sexo, tipo.
  - Cantidad de registros por Localidad.

```python
# dnrpa-inscripciones-iniciales-autos-202201.csv
urlDatasetUno= "https://drive.google.com/uc?id=14Q5P8gJ6iDd6o779HVofh7GxrhJjda-b"
dnrpa = pd.read_csv(urlDatasetUno, sep=";", encoding='iso-8859-1')
```

### <b>DataSet 2:</b> Emisiones de gases en centrales eléctricas - 2018 - EEUU

En el segundo set de datos seleccionado se analizan diferentes equipamientos para control de gases contaminantes en diferentes centrales de generación eléctrica en Estados Unidos durante el año 2018.

Las variables importantes en este dataset son:

- PlantID: identificación de la planta.
- Name of Environmental Equipment or Technology Type: tipo de tecnología o equipamiento para control de emisiones.
- Status: manifiesta si está operativo o no.
- Hours in\n Service: horas en servicio.
- FGD Sorbent Quantity \n(thousand tons): cantidad de absorbentes utilizados.
- FGD Electricity\n Consumption \n(Megawatthours): cantidad de absorbentes utilizados por MWh generado.
- Los porcentajes de control eficiente de los diferentes contaminantes.


```python
#Emisiones EUA2018.csv
urlDatasetDos= "https://drive.google.com/uc?id=1CWOIapptFYcAbk-GN9ZXew88tS4xCyU_"

EUA2018 = pd.read_csv(urlDatasetDos, sep=";")
```

### <b>DataSet 3:</b> Turnos médicos en centros de salud - 2016 - Vitoria ES BR

  - Datos del paciente: id de paciente, sexo, edad, grado de escolarización.

  - Asiste: Indica si el paciente asistió.
  - Centro_de_Salud: Nombre del centro de salud asignado
  - Historia clinica: se indican datos relativos a la historia clínica del paciente como hipertensión, diabetes, alcoholismo.
